**What is the highest value reached by each state at any time of the year in 2019?**

```python
data.head(2)
```

```python
table10=data[data["year"]!=2020]
```

```python
states=list(table10.columns)
states.remove('year')
states.remove('month')
states.remove('day')
```

```python
for i in states:
    x=max(table10[i])
    print(f"the highest value reached by state {i} \t in \t 2019 is \t {x}")
```

```python
dict3=dict()
for i in states:
    x=max(table10[i])
    dict3.update({i:x})
    
table11=pd.DataFrame.from_dict(dict3,orient="index")
table11.rename(columns={0:"highest value reached"},inplace=True)
table11.head(3)
```

**What is the lowest value reached by each state at any time of the year in 2019?**

```python
for i in states:
    x=min(table10[i])
    print(f"the least value reached by state {i} \t in \t 2019 is \t {x}")
```

```python
dict4=dict()
for i in states:
    x=min(table10[i])
    dict4.update({i:x})
    
table12=pd.DataFrame.from_dict(dict4,orient="index")
table12.rename(columns={0:"least value reached"},inplace=True)
table12.head(3)
```
